```python
import numpy as np
import matplotlib.pyplot as plt
import networkx
import networkx.algorithms.approximation



#1

def power_method(A, tol, max):
    n = A.shape[0]
    y = np.random.rand(n)
    #print(n)

    err = 1

    iter = 0
    while err > tol:

        if iter > max:
            break

        y_norma = np.linalg.norm(y)
        y = y / y_norma
        z = A @ y
        z_norm = np.linalg.norm(z)
        z = z / z_norm

        err = abs(1 - abs(np.transpose(z) @ y))
        y = z
        iter += 1

    return y


A = np.random.random((6, 6))
# A = np.array([[6, 5],
#               [4, 5]])

y = power_method(A, 10**(-6), 10000)
print(y)


v, w = np.linalg.eig(A)
# print(v)
print(w[:, 0])


#se observa ca valorile sunt asemanatoare
```

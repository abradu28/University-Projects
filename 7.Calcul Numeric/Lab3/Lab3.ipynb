```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
#1

#a) determinare rang matrice

A = np.random.rand(10, 4)

rank = np.linalg.matrix_rank(A)

print(f'Rangul matricei este: {rank}')
#print(A)
```

```python
#b)

A = np.random.rand(10, 4)
#adaugam primele 4 coloane si la final
A = np.append(A, A[:, 0:4], axis=1)
rank = np.linalg.matrix_rank(A)

#matricea are rang nemaximal < 8

print(f"Rangul noii matrici: {rank}")
#print(A.shape)

#print(A)
```

```python
#c)
zgomot_A = np.random.normal(0,0.2, (10, 8))
print('Rangul matricei cu zgomot este {}'.format(np.linalg.matrix_rank(zgomot_A)))


U, S, V = np.linalg.svd(zgomot_A)
print(f'Valorile singulare ale noii matrici: {S}')

```

```python
#2) comprimare imagine utilizand SVD


A = plt.imread(r'C:\Users\Radu\Desktop\facultate\Anul3\Calcul Numeric\Lab3\pisica.jpg')
plt.imshow(A)



```

```python
#a) Calculare SVD
U, S, V = np.linalg.svd(A)

print(A.shape)


```

```python
#b) alegem k = 15 initial

def aproximare_imagine(k):
    return U[:, :k] @ np.diag(S[:k]) @ V[:k]


plt.imshow(aproximare_imagine(5))
```

```python
#d) Alegem 2 valori care sa fie mai mici decat minimul dintre nr de linii si de coloane

plt.imshow(aproximare_imagine(40))



```

```python
k = 60

plt.imshow(aproximare_imagine(k))

```

Se poate observa ca cu cat este $k$ mai mare, cu atat imaginea este mai clara, aproape la fel ca cea initiala, doar ca numarul operatiilor este mult mai mic.
